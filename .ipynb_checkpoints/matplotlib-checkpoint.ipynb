```python
import matplotlib.pyplot as plt
import numpy as np
import numpy
import math
```

```python
mean = 0
std = 0.2
num_samples = 1000
samples = numpy.random.normal(mean, std, size=num_samples)

plt.plot(samples)
plt.show()
```

```python
x = np.linspace(0,100,1000)
y = 2*np.sin(2*x)-np.cos(3*x)
z = np.cos(x)+np.sin(3*x)+samples
t = np.cos(0.5*x)*np.sin(1.5*x)
```

```python
plt.figure(figsize=(30, 10))
plt.plot(x,y ,'k--')
plt.plot(x,z ,'c')
plt.plot(x,t ,'b--')
plt.xlabel('time')
plt.ylabel('output')
plt.title('random plot')
plt.legend(['first' ,'second', 'third'])
plt.grid(True)
plt.show()
```
