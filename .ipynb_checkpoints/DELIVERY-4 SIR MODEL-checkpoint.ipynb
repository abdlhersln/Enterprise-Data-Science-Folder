Simulative Approach

```python
N0=1000000 #max susceptible population
beta=0.4   # infection spread dynamics
gamma=0.1  # recovery rate

I0=df_plot.Germany[35]
S0=N0-I0
R0=0

def SIR_model(SIR,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return([dS_dt,dI_dt,dR_dt])
```

```python
SIR=np.array([S0,I0,R0])
propagation_rates=pd.DataFrame(columns={'susceptible':S0,
                                        'infected':I0,
                                        'recoverd':R0})


#just simulative approach 100 kez döndür burayi daha sonra diferansiyel denklemler ve parametreler nasil calisiyor izle

for each_t in np.arange(100):
   
    new_delta_vec=SIR_model(SIR,beta,gamma)
   
    SIR=SIR+new_delta_vec
    
    propagation_rates=propagation_rates.append({'susceptible':SIR[0],
                                                'infected':SIR[1],
                                                'recovered':SIR[2]}, ignore_index=True)
```

```python
fig, ax1 = plt.subplots(1, 1)

ax1.plot(propagation_rates.index,propagation_rates.infected,label='infected',color='k')
ax1.plot(propagation_rates.index,propagation_rates.recovered,label='recovered')
ax1.plot(propagation_rates.index,propagation_rates.susceptible,label='susceptible')

ax1.set_ylim(10, 1000000)
ax1.set_yscale('linear')
ax1.set_title('Szenario SIR simulations  (demonstration purposes only)',size=16)
ax1.set_xlabel('time in days',size=16)
ax1.legend(loc='best',
           prop={'size': 16});
```

# Fitting data to a model

```python
def SIR_model_t(SIR,t,beta,gamma):
    ''' Simple SIR model
        S: susceptible population
        t: time step, mandatory for integral.odeint
        I: infected people
        R: recovered people
        beta: 
        
        overall condition is that the sum of changes (differnces) sum up to 0
        dS+dI+dR=0
        S+I+R= N (constant size of population)
    
    '''
    
    S,I,R=SIR
    dS_dt=-beta*S*I/N0          #S*I is the 
    dI_dt=beta*S*I/N0-gamma*I
    dR_dt=gamma*I
    return dS_dt,dI_dt,dR_dt

def fit_odeint(x, beta, gamma):
    '''
    helper function for the integration
    '''
    return integrate.odeint(SIR_model_t, (S0, I0, R0), t, args=(beta, gamma))[:,1] # we only would like to get dI
```

```python
ydata = propagation_rates.infected
t=np.arange(len(ydata))
```

```python
popt, pcov = optimize.curve_fit(fit_odeint, t, ydata)
perr = np.sqrt(np.diag(pcov))
    
print('standard deviation errors : ',str(perr), ' start infect:',ydata[0])
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
```

```python
fitted=fit_odeint(t, *popt)
```

```python
plt.semilogy(t, ydata, 'o')
plt.semilogy(t, fitted)

plt.title("Fit of SIR model for Germany cases")
plt.ylabel("Total Population infected")
plt.xlabel("Days")
plt.show()
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
print("Basic Reproduction Number R0 " , popt[0]/ popt[1])
print("This ratio is derived as the expected number of new infections (these new infections are sometimes called secondary infections from a single infection in a population where all subjects are susceptible. @wiki")
```

# SIR MODELLING

FOR GERMANY

```python
initial_point = 50               #for all countries
last_point = 100+initial_point   #for all countries
ydata = np.array(df_plot.Germany[initial_point:last_point])
t=np.arange(len(ydata))

N0=1000000
beta=0.9
gamma=0.01

I0=ydata[0]
S0=N0-I0
R0=0

popt=[beta,gamma]
# popt=[0.4,0.1]
fit_odeint(t, *popt)


popt, pcov = optimize.curve_fit(fit_odeint, t, ydata)
perr = np.sqrt(np.diag(pcov))
    
print('standard deviation errors : ',str(perr), ' start infect:',ydata[0])
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])

fitted=fit_odeint(t, *popt)

dash_data_germany = pd.DataFrame({'t':t,'germany_data':ydata,'germany_fitted':fitted})

plt.figure(figsize=(18, 7))
plt.semilogy(t, ydata,label='data',color='g')
plt.semilogy(t, fitted,label='fitted',color='k')
plt.legend(loc='best',
           prop={'size': 16});
plt.title("Fit of SIR model for Germany cases",size=16)
plt.ylabel("Population infected")
plt.xlabel("Days",size=16)
plt.show()
print("Optimal parameters: beta =", popt[0], " and gamma = ", popt[1])
print("Basic Reproduction Number R0 " , popt[0]/ popt[1])
print("This ratio is derived as the expected number of new infections (these new infections are sometimes called secondary infections from a single infection in a population where all subjects are susceptible. @wiki")
```

# DASHBOARD

```python
fig3 = go.Figure()
app = dash.Dash()
app.layout = html.Div([
    
    dcc.Markdown('''
    #  Enterprise Data Science Delivery 4 SIR MODELS
    '''),
          
    html.Div([
        "Country Selection",
        dcc.Dropdown( id='country_drop_down3',
        options=[
            {'label': 'Germany', 'value': 'Germany Data'},
            {'label': 'Turkey', 'value': 'Turkey Data'},
            {'label': 'US', 'value': 'US Data'},  
            ],
        value=['Germany Data','Turkey Data'], # which are pre-selected
        multi=True),
         dcc.Graph(figure=fig3, id='main_window_slope3')
             ]),
     
])
```

```python
@app.callback(
    Output('main_window_slope3', 'figure'),
    [Input('country_drop_down3', 'value')])

def update_figure(data_list):

    traces3 = [] 
    for each in data_list:
        
        
        traces3.append(dict (   x=all_data.iloc[:,0],
                                y=real_data[each],
                                mode='markers+lines',
                                opacity=1,
                                line_width=2,
                                marker_size=4, 
                                name=each
                                                                                       
                               
                      )
        )     
        
        traces3.append(dict (   #x=all_data.iloc[:,0],
                                y=fitted_data[each],
                                mode='markers+lines',
                                opacity=1,
                                line_width=2,
                                marker_size=4, 
                                name="SIR Model "+each
                                                                                       
                               
                      )
        )     
        
    return {
            'data': traces3,
            'layout': dict (
                width=1100,
                height=600,
                xaxis_title="Time",
                yaxis_title="SIR Models",
                xaxis={'tickangle':-45,
                        'nticks':20,
                        'tickfont':dict(size=10,color="#7f7f7f"),
                        
                      },
                yaxis={'type':"log",
                       'range':'[1.1,5.5]'
                      }
                         )
            }
```
